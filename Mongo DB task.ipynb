```python
import pandas as pd
from pandas import json_normalize
df=pd.json_normalize(data,'scores',['_id','name'])
#df=pd.json_normalize(data['scores.score'],max_level=2)
```

```python
one=df.loc[df['type']=='exam']
A=one.drop(['type','_id','name'],axis=1)
A.rename(columns={"score":"Exam"},inplace=True)
```

```python
A=A.reset_index(drop=True)
A
```

```python
Two=df.loc[df['type']=='quiz']
B=Two.drop(['type','_id','name'],axis=1)
B.rename(columns={"score":"Quiz"},inplace=True)
```

```python
B=B.reset_index(drop=True)
B
```

```python
Three=df.loc[df['type']=='homework']
C=Three.drop(['type','_id','name'],axis=1)
C.rename(columns={"score":"Homework"},inplace=True)
```

```python
C=C.reset_index(drop=True)
C
```

```python

Scores=pd.concat([A,B,C],axis=1)
```
